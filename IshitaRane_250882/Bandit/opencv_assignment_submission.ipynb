## 1. Reading Images and Videos
**Task:**
1. Read the image `cats.jpg` and display it.
2. Read the video `dog.mp4` and display it frame by frame. Ensure the video window closes when the 'd' key is pressed.

```python
# Write your code here
import cv2 as cv
#Reading and displaying an image 
img=cv.imread('cats.jpg')
if img is None:
    print("Error,pls try again")
else
    cv.imshow('Cats',img)
    cv.waitKey(0)
    cv.destroyAllWindows()


#Reading and displaying a vdo frame by frame
cap = cv.VideoCapture('dog.mp4')
if not cap.isOpened():
    print("Error,pls try again")
else:
    while True:
        ret, frame = cap.read()

        # ret is False when video ends
        if not ret:
            print("Video ended.")
            break

        cv.imshow('Dog Video', frame)

        # press 'd' to exit the video
        if cv.waitKey(25) & 0xFF == ord('d'):
            break

    cap.release()
    cv.destroyAllWindows()


```

## 4. Image Transformations
**Task:**
1. Create a function `translate(img, x, y)` to shift an image.
2. Create a function `rotate(img, angle, rotPoint=None)` to rotate an image.
3. Apply a flip transformation to `park.jpg`.

```python
# Write your code here
import cv2 as cv
import numpy as np
img = cv.imread('park.jpg')
#translation
def translate(img, x, y):
    trans_mat = np.float32([[1, 0, x],
                            [0, 1, y]])
    dimensions = (img.shape[1], img.shape[0])  # (width, height)
    return cv.warpAffine(img, trans_mat, dimensions)
#rotation
def rotate(img, angle, rotPoint=None):
    (height, width) = img.shape[:2]
    if rotPoint is None:
        rotPoint = (width // 2, height // 2)
    rotMat = cv.getRotationMatrix2D(rotPoint, angle, scale=1.0)
    dimensions = (width, height)
    return cv.warpAffine(img, rotMat, dimensions)
# apply translation: shift 100px right, 50px down
translated = translate(img, 100, 50)
cv.imshow('Translated', translated)

# rotate by 45 degrees around center
rotated = rotate(img, 45)
cv.imshow('Rotated 45deg', rotated)

# flip: 0 = vertical flip, 1 = horizontal, -1 = both
flipped_h = cv.flip(img, 1)   # horizontal flip
flipped_v = cv.flip(img, 0)   # vertical flip
cv.imshow('Flipped Horizontal', flipped_h)
cv.imshow('Flipped Vertical', flipped_v)

cv.waitKey(0)
cv.destroyAllWindows()
```

## 8. Bitwise Operations
